# Generators - Practice Solutions

In these exercises, you will create and use generator functions to understand how the `yield` keyword works and why generators are useful for memory efficiency and code modularity.

# Problem 1
Create a generator called `countdown` that yields numbers from a given starting value down to 1.

Then use a `for` loop to print all the values from the generator.

**Sample output**
```
5
4
3
2
1
```

````python
# TODO create the countdown generator
def countdown(num):
  while num > 0:
    yield num
    num -= 1

# Example usage:
for num in countdown(5):
    print(num)
````

# Problem 2
Create a generator called `even_numbers` that yields even numbers from `start` to `end` (inclusive).

Use the `next()` function to manually retrieve the first three even numbers from the generator.

**Sample output**
```
First even: 2
Second even: 4
Third even: 6
```

````python
# TODO create the even_numbers generator
def even_numbers(start, end):
  for num in range(start, end + 1):
    if num % 2 == 0:
      yield num

# Example usage:
# gen = even_numbers(1, 20)
# # Example usage:
gen = even_numbers(1, 20)
print("First even:", next(gen))
print("Second even:", next(gen))
print("Third even:", next(gen))
print("Fourth even: ", next(gen))
````

# Problem 3
Create a generator called `fibonacci` that yields Fibonacci numbers indefinitely (1, 1, 2, 3, 5, 8, 13, ...).

Use this generator to find the first Fibonacci number greater than 100.

**Hint:** Since the generator runs indefinitely, you'll need to break out of your loop once you find the answer.

**Sample output**
```
The first Fibonacci number greater than 100 is: 144
```

````python
# TODO create the fibonacci generator and find the first Fibonacci number > 100
def fibonacci():
  a, b = 0, 1
  while True:
    yield a
    a, b = b, a+b

# Example usage:
for fib in fibonacci():
    if fib > 150:
        print(f"The first Fibonacci number greater than 100 is: {fib}")
        break
````
